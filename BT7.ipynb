```python
#W7A1
def binary_search(arr,left,right,target):
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==target:return mid
        if arr[mid]<target:left=mid+1
        else:right=mid-1
    return -1
a=list(map(int,input().split()))
b=int(input())  
result=binary_search(a,0,len(a)-1,b)
if result!=-1:
    print(str(result)) 
else:
    print("-1")        
```

```python
#W7A2
def count_occurrences(arr,x):
    c=0
    for i in arr:
        if i==x:c+=1
        return c
a=list(map(int,input().split()))
b=int(input())  
result=count_occurrences(a,b)
print(str(result))
```

```python
#W7A3
def bubble_sort(a):
    n=len(a)
    for i in range(n):
        for j in range(0,n-i-1):
            if a[j]>a[j+1]:a[j],a[j+1]=a[j+1],a[j]
    return a
def selection_sort(a):
    n=len(a)
    for i in range(n):
        m=i
    for j in range(i+1,n):
        if a[j]<a[m]:m=j
        a[i],a[m]=a[m],a[i]
    return a
a=list(map(int,input().split()))
b=int(input())  
if b==1:             
    result=bubble_sort(a)
else:
    result=selection_sort(a)
print(' '.join(map(str,result)))
```

```python
#W7A4
def most_frequent(arr):
    d={}
    for i in arr:
        d[i]=d.get(i,0)+1
        mx=max(d.values())
    for i in arr:
        if d[i]==mx:return i,mx
a=list(map(int,input().split()))
num,freq=most_frequent(a)
print(f"{num} {freq}")      
```

```python
#W7A5
def count_pairs(arr,x):
 c=0
 for i in range(len(arr)):
  for j in range(i+1,len(arr)):
   if arr[i]+arr[j]==x:c+=1
 return c
a=list(map(int,input().split()))    
b=int(input())
result=count_pairs(a,b)
print(str(result))
```

```python
#W7A6
def flatten(lst):
 r=[]
 for i in lst:
  for j in i:r.append(j)
 return r
lst=[]
n=int(input())
for i in range(n):
 lst.append(list(map(int,input().split())))
result=flatten(lst)
print(' '.join(map(str,result)))    
```

```python
#W7A7
def lis(nums):
 dp=[1]*len(nums)
 for i in range(len(nums)):
  for j in range(i):
   if nums[j]<nums[i]:dp[i]=max(dp[i],dp[j]+1)
 return max(dp) if nums else 0
a=list(map(int,input().split()))
result=lis(a)
print(str(result))  
```

```python
#W7A8
def min_interval(intervals,queries):
 res=[]
 for q in queries:
  best=10**18
  for l,r in intervals:
   if l<=q<=r:best=min(best,r-l+1)
  res.append(best if best!=10**18 else -1)
 return res
intervals=[]    
n=int(input())
for i in range(n):
 intervals.append(list(map(int,input().split())))
q=int(input())
queries=[]
for i in range(q):
 queries.append(int(input()))
result=min_interval(intervals,queries)      
```

```python
#W7A9
def count_smaller(nums):
 r=[]
 for i in range(len(nums)):
  c=0
  for j in range(i+1,len(nums)):
   if nums[j]<nums[i]:c+=1
  r.append(c)
 return r
a=list(map(int,input().split()))
result=count_smaller(a)
print(' '.join(map(str,result)))        
```

```python
#W7A10
def max_subarray_mod(nums,m):
 pre=0
 best=0
 s=[]
 import bisect
 for x in nums:
  pre=(pre+x)%m
  i=bisect.bisect_right(s,pre)
  if i<len(s):best=max(best,(pre-s[i]+m)%m)
  else:best=max(best,pre)
  bisect.insort(s,pre)
 return best
a=list(map(int,input().split()))
m=int(input())
result=max_subarray_mod(a,m)
print(str(result))  
```

```python
#W7A11
def count_word(s,w):
 c=0
 i=0
 while True:
  i=s.find(w,i)
  if i==-1:break
  c+=1;i+=len(w)
 return c
s=input()
w=input()
result=count_word(s,w)
print(str(result))  
```

```python
#W7A13
def harmonious(arr):
 d={}
 for i in arr:d[i]=d.get(i,0)+1
 ans=0
 for k in d:
  if k+1 in d:ans=max(ans,d[k]+d[k+1])
 return ans
a=list(map(int,input().split()))
result=harmonious(a)
print(str(result))      
```
