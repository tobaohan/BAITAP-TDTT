```python
#w6a5
n = int(input())
a=[]
cnt={"positive":0,"negative":0,"zero":0}
for i in range(n):
    a.append(int(input()))
    if a[i]>0:
        cnt["positive"]+=1
    elif a[i]<0:
        cnt["negative"]+=1
    else:
        cnt["zero"]+=1
print("Positive:",cnt["positive"])
print("Negative:",cnt["negative"])  
print("Zero:",cnt["zero"])
```

```python
#w6a9
import ast

s1 = input()
s2 = input()
d1 = ast.literal_eval(s1)
d2 = ast.literal_eval(s2)
cnt = {}
for i in d1:
    if i not in cnt:
        cnt[i]=0
    cnt[i]+=d1[i]
for i in d2:
    if i not in cnt:
        cnt[i]=0
    cnt[i]+=d2[i]
print(cnt)

```

```python
#w6a10
a=list(map(int, input().split()))
k=int(input())
kq=[]
for i in range(len(a)):
    for j in range(i+1, len(a)):
        if a[i]+a[j]==k:
            b=(a[i],a[j])
            kq.append(b)
print(kq)
```

```python
#w6a12
a=list(map(int, input().split()))
max=0
imax=1e9
cnt={}
for i in a:
    if i not in cnt:
        cnt[i]=0
    cnt[i]+=1
    if cnt[i]>max:
        max=cnt[i]
        imax=i
    elif cnt[i]==max:
        if i<imax:
            max=cnt[i]
            imax=i
print(imax)
        
```
