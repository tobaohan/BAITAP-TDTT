```python
#w7a16
a=list(map(int, input().split()))
a.sort()
cnt={}
for i in range(len(a)):
    if a[i] not in cnt:
        cnt[a[i]]=1
    else:
        cnt[a[i]]+=1
ktra=True
them=0
for k in cnt:
    if cnt[k]<k:
        them+=cnt[k]
        ktra=False
    elif cnt[k]>k:
        them+=cnt[k]-k
        ktra=False
if ktra:
    print("YES")
else:
    print("NO")
    print(them)
```

```python
#W7A17
a=list(map(int,input().split()))
cand=0
cnt=0
for x in a:
	if cnt==0:
		cand=x;cnt=1
	elif x==cand:
		cnt+=1
	else:
		cnt-=1
if a.count(cand)>len(a)//2:
	print(cand)
```

```python
#W7A19
a=list(map(int,input().split()))
b=list(map(int,input().split()))
i=j=0
res=[]
inc=a[0]<=a[-1]
while i<len(a) and j<len(b):
	if (a[i]<=b[j])==inc:
		res.append(a[i]);i+=1
	else:
		res.append(b[j]);j+=1
res+=a[i:]+b[j:]
print(*res)
```

```python
#W7A22
n=int(input())
cnt={}
for _ in range(n):
	name=input().strip().split()[-1]
	cnt[name]=cnt.get(name,0)+1
mx=max(cnt.values())
res=[k for k in cnt if cnt[k]==mx]
print(sorted(res)[0])
```

```python
#W7A25
s=input().strip()
k=int(input())
st=[]
for i,x in enumerate(s):
	while st and k>0 and st[-1]<x:
		st.pop();k-=1
	st.append(x)
print("".join(st[:len(s)-k]))
```

```python
#W7A27
s=input()
st=[]
ok=True
mp={')':'(','}':'{',']':'[','>':'<'}
for c in s:
	if c in '({[<':
		st.append(c)
	else:
		if not st or st.pop()!=mp.get(c,''):
			ok=False;break
print("YES" if ok and not st else "NO")
```

```python
#W7A28
n=int(input())
res=[]
cur=[]
i=4
while i<=n:
	x=i
	if x>3:
		cur.append(x)
	else:
		if len(cur)>len(res):
			res=cur[:]
		cur=[]
	i+=1
if len(cur)>len(res):
	res=cur
print(", ".join(map(str,res)))
```

```python
#W7A29
a=list(map(int,input().split()))
best=cur=[a[0]]
for i in range(1,len(a)):
	if a[i]>a[i-1]:
		cur.append(a[i])
	else:
		if len(cur)>len(best):
			best=cur[:]
		cur=[a[i]]
if len(cur)>len(best):
	best=cur
print(*best)
```

```python
#W7A30
a=list(map(int,input().split()))
dp=[1]*len(a)
par=[-1]*len(a)
for i in range(len(a)):
	for j in range(i):
		if a[j]<a[i] and dp[j]+1>dp[i]:
			dp[i]=dp[j]+1
			par[i]=j
m=max(dp);idx=dp.index(m)
res=[]
while idx!=-1:
	res.append(a[idx])
	idx=par[idx]
print(*res[::-1])
```
