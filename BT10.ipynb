```python
#w9a1
a=list(map(int, input().split()))
b=list(map(int, input().split()))
k=0
c=[]
for i in range(len(a)):
    if a[i]<=b[k]:
        c.append(a[i])
    else:
        c.append(b[k])
        k+=1
while k<len(b):
    c.append(b[k])
    k+=1
for i in c:
    print(i, end=" ")
```

```python
#w9a2
n=int(input())
m=int(input())
k=int(input())
a=[]
for i in range(n):
    row=[]
    for j in range(m):
        row.append(int(input()))
    a.append(row)
a.sort()
rs=[]
for i in range(n):
    for j in range(m):
        rs.append(a[i][j])
print(a)
print(rs)
print(rs[k-1])
```
