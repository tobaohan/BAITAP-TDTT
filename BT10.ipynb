```python
#w9a1
a=list(map(int, input().split()))
b=list(map(int, input().split()))
k=0
c=[]
for i in range(len(a)):
    if a[i]<=b[k]:
        c.append(a[i])
    else:
        c.append(b[k])
        k+=1
while k<len(b):
    c.append(b[k])
    k+=1
for i in c:
    print(i, end=" ")
```

```python
n = int(input())
a = list(map(int, input().split()))
m = int(input())
b = list(map(int, input().split()))
si_so = []
tong = 0
for i in range(len(a)):
  tong += a[i]
  si_so.append(tong)

ket_qua = []
for i in range(len(b)):
  for j in range(len(si_so)):
    if b[i] <= si_so[j]:
      ket_qua.append(j+1)
      break
print(*ket_qua)
```

```python
import math
n,m = map(int,input().split())
a = list(map(int, input().split()))
b = list(map(int, input().split()))
uc = []
for i in range(len(a)):
  for j in range(len(b)):
    c = math.gcd(a[i],b[j])
    if c not in uc:
      uc.append(c)
print(*uc)
```
