```python
#w5a3
from math import *
def prime(n):
    if n<2:
        return False
    else:
        for i in range(2, ceil(sqrt(n))+1):
            if n%i==0:
                return False
        return True
n = int(input())
if prime(n):
    print("YES")
else: 
    print("NO")
```

```python
def anhxa(a, b):
    f = []
    fx = []
    cnt = {}
    for i in range(len(a)):
        if a[i] not in f and a[i]!=b[i]:
            f.append(a[i])
            fx.append(b[i])
    for i in range(len(fx)):
        if fx[i] not in cnt:
            cnt[fx[i]]=0
        cnt[fx[i]] += 1
        if cnt[fx[i]] > 1:
            return False
    for i in range(len(a)):
        for j in range(len(f)):
            if a[i] == f[j]:
                if b[i] != fx[j] and a[i] != b[i]:
                    return False
    return True

a, b = input().split()
if anhxa(a, b):
    print("True")
else:
    print("False")



```

```python
s = input()
check=[None]*len(s)
for i in range(len(s)):
    check[i]=False
for i in range(len(s)):
    for j in range(i-1, -1, -1):
        if s[i]==")":
            if s[j]=="(" and check[j]==False:
                check[j]=True
                check[i]=True
        elif s[i]=="]":
            if s[j]=="[" and check[j]==False:
                check[j]=True
                check[i]=True
        elif s[i]=="}":
            if s[j]=="{" and check[j]==False:
                check[j]=True
                check[i]=True
ktra=0
for i in range(len(check)):
    if check[i]==False:
        print("false")
        ktra=1
        break
if ktra==0:
    print("true")
```
