```python
#w11a1
with open('data.txt', 'w', encoding='utf-8') as f:
  f.write('123456')
def reverse(path:str):
  with open(path, 'r') as f:
    number_str = f.read().strip()
    return (number_str[::-1])
print(reverse('data.txt'))
```

```python
#w11a2
with open('data.txt', 'w', encoding='utf-8') as f:
    f.write('7')
def canWinNim(path: str) -> bool:
    with open(path, "r") as f:
        n = int(f.read().strip())
    return n % 4 != 0
print(canWinNim('data.txt'))
```

```python
#w11a4
with open('data.txt','w',encoding='utf-8') as f1, \
    open('anagram.txt','w',encoding='utf-8') as f2:
            f1.write("Tom Marvolo Riddle")
            f2.write("I am Lord Voldemort")
def are_anagrams(path1: str, path2: str):
      with open(path1, 'r', encoding='utf-8') as f1, \
           open(path2, 'r', encoding='utf-8') as f2:
          str1 = f1.read().replace(" ","").lower()
          str2 = f2.read().replace(" ","").lower()
          return sorted(str1) == sorted(str2)
print(are_anagrams('data.txt', 'anagram.txt'))
```

```python
#w11a5
with open('data.txt','w',encoding='utf-8') as f:
    f.write("hello")
def count(path: str, char: str):
    with open(path,'r',encoding='utf-8') as f:
        content = f.read()
        return content.count(char)
print(count('data.txt','l'))
```
