```python
#W11A8:
def write_fibonacci(path):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            n = int(f.readline().split()[0])
            a = int(f.readline().split()[0])
            b = int(f.readline().split()[0])
    except FileNotFoundError:
        print("Không tìm thấy tệp tin!")
        return

    fib = [0] * (n + 1)
    if n >= 0: fib[0] = a
    if n >= 1: fib[1] = b

    for i in range(2, n + 1):
        fib[i] = fib[i-1] + fib[i-2]

    try:
        with open('output.txt', 'w', encoding='utf-8') as f:
            for num in fib:
                f.write(str(num) + "\n")
        print("Đã ghi file thành công.")
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")

write_fibonacci('input.txt')
with open('output.txt', 'r', encoding='utf-8') as f:
  content = f.read()
  print(content)
```

```python
#W11A10:
def getMoney(path: str) -> list[int]:
    try:
        with open(path, 'r', encoding='utf-8') as f:
            content = f.read()
            money_list = [int(x) for x in content.split()]
            return money_list
    except FileNotFoundError:
        print("Không tìm thấy tệp tin!")
        return []
    except Exception as e:
        print(f"Lỗi: {e}")
        return []

def rob(houses: list[int]) -> int:
    if not houses:
        return 0
    if len(houses) == 1:
        return houses[0]

    prev2 = 0
    prev1 = 0

    for amount in houses:
        current = max(prev1, prev2 + amount)

        prev2 = prev1
        prev1 = current

    return prev1

with open('money.txt', 'w') as f:
    f.write("10 20 30 1 5")

money_in_houses = getMoney('money.txt')
max_money = rob(money_in_houses)

print(f"Danh sách số tiền trong các nhà: {money_in_houses}")
print(f"Số tiền tối đa có thể trộm được: {max_money}")
```

```python
#W11A12:
def maximumProduct(path: str) -> int:
    try:
        with open(path, 'r') as f:
            data = f.read().split()
            numbers = [int(x) for x in data[1:]]

            numbers.sort()

            option1 = numbers[-1] * numbers[-2] * numbers[-3]
            option2 = numbers[0] * numbers[1] * numbers[-1]

            return max(option1, option2)
    except:
        return 0
print(maximumProduct('input.txt'))
```

```python
#W11A14:
def averageTime(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        if len(lines) <= 1:
            return 0.0

        total_time = 0.0
        count = 0

        for line in lines[1:]:
            line = line.strip()
            if not line:
                continue

            parts = line.split(',')

            time_value = float(parts[1])
            total_time += time_value
            count += 1

        if count == 0:
            return 0.0

        return round(total_time / count, 2)

    except (FileNotFoundError, ValueError, IndexError):
        return 0
print(averageTime('du_lieu_dua.csv'))
```

```python
#W11A15:
def findMovies(path:str):
  try:
    with open(path, 'r', encoding = 'utf-8') as f:
      lines = f.readlines()
    result = []
    for line in lines[1:]:
      line = line.strip()
      if not line:
        continue

      parts = line.split(',')

      if len(parts) >= 4:
        if float(parts[3]) >= 8.0:
          result.append(parts[1])
    if result:
      return result
    else:
      return 'All are boring'
  except FileNotFoundError:
    print("Không tìm thấy tệp tin!")
findMovies('/content/findMovies.csv')
```
