```python
#W8A1
class Cylinder:
	def __init__(self,r,h):
		self.r=r
		self.h=h
	def surface(self):
		return 2*3.14*self.r*(self.r+self.h)
	def volume(self):
		return 3.14*self.r*self.r*self.h
r,h=map(float,input().split())
c=Cylinder(r,h)
print(f"{c.surface():.2f} {c.volume():.2f}")
```

```python
#W8A2
class Date:
	def __init__(self,d,m,y):
		self.d=d;self.m=m;self.y=y
	def nhuan(self):
		return self.y%400==0 or (self.y%4==0 and self.y%100!=0)
	def valid(self):
		md=[31,29 if self.nhuan() else 28,31,30,31,30,31,31,30,31,30,31]
		return 1<=self.m<=12 and 1<=self.d<=md[self.m-1]
	def next(self):
		md=[31,29 if self.nhuan() else 28,31,30,31,30,31,31,30,31,30,31]
		self.d+=1
		if self.d>md[self.m-1]:
			self.d=1;self.m+=1
			if self.m>12:self.m=1;self.y+=1
s=input()
d,m,y=map(int,s.split("/"))
dt=Date(d,m,y)
if not dt.valid():print("INVALID")
else:
	dt.next()
	print(f"{dt.d:02d}/{dt.m:02d}/{dt.y:04d}")
```

```python
#W8A3
import math
class Point:
	def __init__(self,x,y):
		self.x=x;self.y=y
	def origin(self):
		return self.x==0 and self.y==0
	def on_x(self):
		return self.y==0
	def on_y(self):
		return self.x==0
	def dist(self):
		return math.sqrt(self.x*self.x+self.y*self.y)
x,y=map(float,input().split())
p=Point(x,y)
if p.origin():print("tren goc toa do")
elif p.on_y():print("tren truc y")
elif p.on_x():print("tren truc x")
else:print("ko tren truc toa do")
print(f"{p.dist():.2f}")
```

```python
#W8A5
import math
class Quadratic:
	def __init__(self,a,b,c):
		self.a=a;self.b=b;self.c=c
	def solve(self):
		if self.a==0:
			if self.b==0:return "No solution"
			return f"{-self.c/self.b:.2f}"
		d=self.b*self.b-4*self.a*self.c
		if d<0:return "No solution"
		if d==0:return f"{-self.b/(2*self.a):.2f}"
		x1=(-self.b+math.sqrt(d))/(2*self.a)
		x2=(-self.b-math.sqrt(d))/(2*self.a)
		return f"{x1:.2f} {x2:.2f}"
a,b,c=map(float,input().split())
print(Quadratic(a,b,c).solve())
```

```python
#W8A8
class LinearSystem:
	def __init__(self,a,b,c,d,m,n):
		self.a=a;self.b=b;self.c=c;self.d=d;self.m=m;self.n=n
	def solve(self):
		D=self.a*self.d-self.b*self.c
		if D==0:return "No solution"
		x=(self.m*self.d-self.b*self.n)/D
		y=(self.a*self.n-self.m*self.c)/D
		return f"{x:.2f} {y:.2f}"
a,b,m=map(float,input().split())
c,d,n=map(float,input().split())
print(LinearSystem(a,b,c,d,m,n).solve())
```

```python
#W8A9
class Time:
	def __init__(self,x):
		self.x=x
	def convert(self):
		h=self.x//3600
		self.x%=3600
		m=self.x//60
		s=self.x%60
		return h,m,s
x=int(input())
t=Time(x)
h,m,s=t.convert()
print(h,m,s)
```

```python
#W8A10
import math
class Circle:
	def __init__(self,x,y,r):
		self.x=x;self.y=y;self.r=r
	def contain(self,a,b):
		return math.sqrt((a-self.x)**2+(b-self.y)**2)<=self.r
x,y,r=map(float,input().split())
a,b=map(float,input().split())
c=Circle(x,y,r)
print("YES" if c.contain(a,b) else "NO")
```

```python
#W8A11
import math
class PlaneAngle:
	def __init__(self,A,B,C,D):
		self.A=A;self.B=B;self.C=C;self.D=D
	def vec(self,P,Q):
		return (Q[0]-P[0],Q[1]-P[1],Q[2]-P[2])
	def cross(self,u,v):
		return (u[1]*v[2]-u[2]*v[1],u[2]*v[0]-u[0]*v[2],u[0]*v[1]-u[1]*v[0])
	def dot(self,u,v):
		return u[0]*v[0]+u[1]*v[1]+u[2]*v[2]
	def norm(self,u):
		return math.sqrt(self.dot(u,u))
	def angle(self):
		n1=self.cross(self.vec(self.A,self.B),self.vec(self.A,self.C))
		n2=self.cross(self.vec(self.B,self.C),self.vec(self.B,self.D))
		return math.degrees(math.acos(self.dot(n1,n2)/(self.norm(n1)*self.norm(n2))))
A=tuple(map(float,input().split()))
B=tuple(map(float,input().split()))
C=tuple(map(float,input().split()))
D=tuple(map(float,input().split()))
print(f"{PlaneAngle(A,B,C,D).angle():.2f}")
```

```python
#W8A12
class ComplexNumber:
	def __init__(self,r,i):
		self.r=r;self.i=i
	def multiply(self,o):
		return ComplexNumber(self.r*o.r-self.i*o.i,self.r*o.i+self.i*o.r)
a,b=map(float,input().split())
c,d=map(float,input().split())
z=ComplexNumber(a,b).multiply(ComplexNumber(c,d))
print(int(z.r),int(z.i))
```

```python
#W8A13
import math
class Triangle:
	def __init__(self,a,b,c):
		self.a=a;self.b=b;self.c=c
	def area(self):
		if self.a+self.b<=self.c or self.a+self.c<=self.b or self.b+self.c<=self.a:
			return -1
		p=(self.a+self.b+self.c)/2
		return math.sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))
a,b,c=map(int,input().split())
t=Triangle(a,b,c)
s=t.area()
if s<0:print("invalid")
else:print(f"{s:.2f}")
```
